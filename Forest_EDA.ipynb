### Checking for outliers

```python
plt.style.use('seaborn')
fig,ax=plt.subplots(nrows=2,ncols=5,figsize=(20,10))
columns_1=np.array([['Temperature','RH','Ws','Rain','FFMC',],['DMC','DC','ISI','BUI','FWI']])
for i in range(0,2):
    for j in range(0,5):
        sns.boxplot(data=df[columns_1[i,j]],ax=ax[i,j])
        ax[i,j].set_title(columns_1[i,j])
```

```python
sns.countplot(data=df,x='month',hue='Class',palette='Accent_r',saturation=0.5,dodge=True)
plt.title('Monthly wise fire analysis')
plt.legend(['Not_fire','fire'])
plt.show()
```
